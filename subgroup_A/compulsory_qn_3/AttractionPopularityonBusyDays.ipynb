```python
# Import necessary libraries
import pandas as pd

# Load the Excel file
file_path = '/content/Survey_cleaned_balanced.xlsx'  # Adjust the path if needed
data = pd.read_excel(file_path)

# Find columns containing "attractions busy day"
busy_day_columns = [col for col in data.columns if 'attractions busy day' in col.lower()]

# Extract only the relevant columns for popularity analysis
attractions_busy_day_data = data[busy_day_columns]

# Calculate the average popularity for each attraction
popularity_ranking = attractions_busy_day_data.mean().sort_values(ascending=False)

# Display the results
print("Attractions ranked by popularity on busy days:")
print(popularity_ranking)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '/content/Survey_cleaned_balanced.xlsx'  # Adjust the path if needed
data = pd.read_excel(file_path)

# Find columns containing "attractions busy day"
busy_day_columns = [col for col in data.columns if 'attractions busy day' in col.lower()]

# Extract only the relevant columns for popularity analysis
attractions_busy_day_data = data[busy_day_columns]

# Calculate the average popularity for each attraction
popularity_ranking = attractions_busy_day_data.mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 8))
popularity_ranking.plot(kind='bar', color='skyblue')
plt.title('Attraction Popularity on Busy Days')
plt.xlabel('Attractions')
plt.ylabel('Average Popularity')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
